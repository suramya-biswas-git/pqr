```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
filepath = 'student_data.csv'
studentDf=pd.read_csv(filepath)
studentDf.dropna(inplace=True)
studentDf.drop_duplicates(inplace=True)
studentDf['Major'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title("Students percentage based on Major subjects")
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
filepath='customer_new.csv'
customerDf=pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
customerDf['Payment_method'].value_counts().plot(kind='pie',autopct='%1.0f%%')
plt.title('Payment Method usage')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
filepath='customer_new.csv'
customerDf=pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
customerDf['Amount_spent'].groupby(customerDf['Payment_method']).sum().plot(kind='bar')
plt.xlabel('Payment Mode')
plt.ylabel('Amount Spent')
plt.title('Usage of different Payment mode & Their Amount Spent')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
filepath='customer_new.csv'
customerDf=pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
customerDf['Amount_spent'].groupby(customerDf['Employees_status']).sum().plot(kind='barh')
plt.ylabel('Employee Status')
plt.xlabel('Amount Spent')
plt.title('Different Employee Status &  Amount Spent')
plt.show()
```
